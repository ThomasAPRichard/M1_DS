# Fonction scraping url

```python
def scraping_url(liste_brands):

    liste_brands_name = liste_brands['brand']
    liste_urls_marques = liste_brands['url']


    # ✅ Configuration Selenium
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Désactiver pour voir l'exécution
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36")

    data = []

    for url, brand in zip(liste_urls_marques, liste_brands_name):
        
        # ✅ Initialiser WebDriver
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
        driver.get(f'{url}')
    
        # Récupérer le HTML après exécution de JavaScript
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        driver.quit()

        # Sélectionne tous les blocs contenant les infos d'un parfum
        perfume_blocks = soup.select('div.cell.text-left.prefumeHbox.px1-box-shadow')

        for block in perfume_blocks:
            # Dans chaque bloc, on cherche le <h3> contenant <a href=...>
            a_tag = block.select_one('div.flex-child-auto h3 a')
            if a_tag:
                # Récupère l'URL
                link = a_tag.get('href', None)
                link = 'https://www.fragrantica.com' + link
                # Récupère le texte, ex: "AMG Black Thrill"
                text = a_tag.get_text(strip=True)
                
                # On stocke le résultat dans une structure
                data.append({
                    "brand" : brand,
                    "name": text,
                    "url": link
                })


    df = pd.DataFrame(data)

    return df
```
