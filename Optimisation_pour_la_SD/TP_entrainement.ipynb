```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#TP1

#2.1
n=10
R_theorique=2
sigma = 1
I = np.arange(0,n)
U = R_theorique*I + np.random.randn(n)*sigma

R_opt = (np.dot(U,I))/(np.dot(I,I))
print(R_opt)
```

```python
#2.2
n=10
sigma = 1

a = np.random.randint(0,20,6) #vecteur de R6
X=np.arange(n) # n valeur de X
eps = np.random.randn(n)*sigma #bruit

Y = np.zeros(n)
for i in range (n):
    for j in range(6) :
        Y[i] += a[j]*(X[i]**j)
    Y[i] += eps[i]

V = [[xi**k for k in range(6)] for xi in X]
V = np.array(V)
a_opt = np.dot(np.linalg.inv(V.T@V), V.T@Y)


Y_pred = V@a_opt
print(f"Y ={Y}")
print(f"Y_pred ={Y_pred}")

plt.plot(X,Y,label='Données bruitées')
plt.plot(X,Y_pred, label='Modèle')
plt.legend()
plt.show()

from sklearn.metrics import mean_squared_error as mse
print (mse (Y,Y_pred))
```
